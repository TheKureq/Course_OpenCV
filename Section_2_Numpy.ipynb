```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from PIL import Image
```

```python
pic = Image.open('./img/puppy.jpg')
type(pic)
```

```python
pic_arr = np.asarray(pic)
type(pic_arr)
```

```python
pic_arr.shape
```

```python
plt.imshow(pic_arr)
```

```python
pic_red = pic_arr.copy()
plt.imshow(pic_red)
```

```python
pic_red.shape
```

```python
plt.imshow(pic_red[:,:,2])
```

```python
# RED CHANNEL VALUSE
plt.imshow(pic_red[:,:,0],cmap='gray')
```

```python
pic_red[:,:,0] = 0
pic_red[:,:,1] = 0
plt.imshow(pic_red)
```
